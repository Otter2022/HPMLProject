```python
import matplotlib.pyplot as plt

runtimes = [.3671, .3248, .3198]

speedup = [runtimes[0]/runtimes[0], runtimes[0]/runtimes[1], runtimes[0]/runtimes[2]]

plt.plot([1, 2, 3], speedup, 'ro-') 
plt.xlabel('Number of threads')
plt.ylabel('Speedup')
plt.title('Speedup of parallelized code')
plt.show()
```
